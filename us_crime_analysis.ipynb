# Data Description

US Crime Rate dataset describes various crimes and its count value from year 1960 to 2014

# Attribute Information :

1). Year --> from 1960 to 2014 (numerical values)

2). Population --> Population count at particular Year (numerical values)

3). Total --> Crime Total from the population (numerical values)

4). Violent --> Violent crime count (numerical values)

5). Property --> Property fraudant crime count (numerical values)

6). Murder --> Violent murder crime count (numerical values)

7). Forcible_Rape --> Violent rapes crime count (numerical values)

8). Robbery --> Robbery crime count (numerical values)

9). Aggravated_assault --> Aggravated assualt crime count (numerical values)

10). Burglary --> Burglary crime count (numerical values)

11). Larcency_Theft --> Larcency Theft count (numerical values)

12). Vehicle_Theft --> Vehicle Theft count (numerical values)



```
import pandas as pd
import numpy as np
```

```
url = 'https://raw.githubusercontent.com/CSiddharthNavaneethan/data_analysis/master/US_Crime_Rates_1960_2014.csv'
```

```
# read the file
df = pd.read_csv(url)
```

```
# data description
df.info()

# total rows
df.shape[0]
```

```
df.head()
```

```
# checking for violent crimes
data1 = pd.DataFrame(df)
```

```
# violent crimes sum total
data1['Violent_Crimes_Total'] = data1[['Violent','Murder','Forcible_Rape']].apply(np.sum,axis=1)
```

```
# total violent crime as by Year 
pd.pivot_table(data1,index='Year' 
               ,values=['Violent','Murder','Forcible_Rape','Violent_Crimes_Total']
               ,aggfunc=np.sum
               ,margins_name='Total'
               ,margins=True)
```

```
# create a new dataframe for other violent crimes 
data2 = pd.DataFrame(df)
```

```
# other violent crimes sum total
data2['Violent_Fraudant_and_Other_Crime_Total'] = data2[['Property','Robbery','Aggravated_assault','Burglary','Larceny_Theft','Vehicle_Theft']].apply(np.sum,axis=1)
```

```
# total other violent crimes as by Year
pd.pivot_table(data2,index='Year'
               ,values=['Property','Robbery','Aggravated_assault','Burglary','Larceny_Theft','Vehicle_Theft','Violent_Fraudant_and_Other_Crime_Total']
               ,aggfunc=np.sum
               ,margins_name='Total'
               ,margins=True)
```

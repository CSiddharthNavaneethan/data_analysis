
# Data Description

Drink dataset describe the servings of alcohol in countries

# Attribute Information :

1). country as countries names around the world (String values)

2). beer_servings the quantity of beer served in the countries (numerical values)

3). spirit_servings the quantity of spirit served in the countries (numerical values)

4). wine_servings the quantity of wine served in the countries (numerical values)

5). total_litres_of_pure_alcohol details percentage of alcohol level in the countries served

6). continent as regions around the world  



```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```
df = pd.read_csv('https://raw.githubusercontent.com/CSiddharthNavaneethan/data_analysis/master/drinks.csv')
```

```
df.head()
```

```
df.info()

df.shape[0]
```

```
# checking for particular beer servings in the countries
data1 = pd.DataFrame(df)
```

```
# drop the spirit and wine servings from the dataset
data1.drop(columns=['spirit_servings','wine_servings'],axis=1,inplace=True)
```

```
# high beer servings in countries
beer = data1.groupby(['country','continent'])
beer = beer.sum()
beer = beer.sort_values(by=['beer_servings'],ascending=False)
beer.head(10)
```

```
# total pure high alcohol levels in beer_servings
beer.sort_values(by='total_litres_of_pure_alcohol',ascending=False).head(10)
```

```
# checking for particular spirit servings in the countries
data2 = pd.DataFrame(df)
```

```
# drop the beer and wine serving from the dataset
data2.drop(['beer_servings','wine_servings'],axis=1,inplace=True)
```

```
# high spirit servings in countries
sprt = data2.groupby(['country','continent'])
sprt = sprt.sum()
sprt = sprt.sort_values(by=['spirit_servings'],ascending=False)
sprt.head(10)
```

```
# total pure high alcohol levels in spirit_servings
sprt.sort_values(by=['total_litres_of_pure_alcohol'],ascending=False).head(10)
```

```
# checking for particular wine servings in the countries
data3 = pd.DataFrame(df)
```

```
# drop the beer and spirit servings from the dataset
data3.drop(columns=['beer_servings','spirit_servings'],axis=1,inplace=True)
```

```
# high wine servings in countries
wine = data3.groupby(['country','continent'])
wine = wine.sum()
wine.sort_values(by=['wine_servings'],ascending=False)
wine.head(10)
```

```
# total pure high alcohol level in wine_servings
wine.sort_values(by=['total_litres_of_pure_alcohol'],ascending=False).head(10)
```
